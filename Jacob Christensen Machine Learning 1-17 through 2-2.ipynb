## Excersize 1: Numpy standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create an array with 100 points, spaced equally from 0 to 50
a = np.linspace(0,50,100)
#print(a)
```

```python
# calculate standard deviation with numpy algorithm
np.std(a)
```

```python
# calculate standard deviation with estimator

# mean of A
Neq = (99-0)+1
mean_est = (1/Neq)*np.sum(a)
print(mean_est)
```

```python
# now for std deviation estimate

std_est = ((1/(Neq-1)) * np.sum((a - mean_est)**2))**(1/2)
print(std_est)
```

```python
# we see how this gives a slightly larger value than numpy algorithm
std_est/np.std(a)
```

```python
# let's see how this compares to the actual standard deviation
mean_true = (1/100)*np.sum(a)
std_true = ((1/100)*np.sum((a-mean_true)**2))**(1/2)
print(std_true)
```

```python
# these are the same now since we are using the exact formulas
std_true/np.std(a)
```

# Excersize: Compute RMSE

```python
# create an array filled with the mean value of array b from the previous example
# array b is an array filled with random values from uniform distribution
b = np.random.rand(10000)
mean_array = np.full(10000,np.mean(b))
```

```python
# now compute the RMSE using the function given in the slides
full_RMSE = np.sqrt((1/10000)*(np.sum((b - mean_array)**2)))
print(full_RMSE)
```

```python
# now shuffle the data
np.random.shuffle(b)
print(b)
```

```python
# and now split dataset into three sections
10000/3
```

```python
section1 = b[0:3333]
section2 = b[3333:6667]
section3 = b[6667:10000]
```

```python
# and calculate the RMSE based on the mean of each section
section1_mean = np.mean(section1)
section2_mean = np.mean(section2)
section3_mean = np.mean(section3)

# can also calculate means a little more manually...
# section1_mean = (1/len(section1)) * np.sum(section1)
# section2_mean = (1/len(section2)) * np.sum(section2)
# section3_mean = (1/len(section3)) * np.sum(section3)

section1_RMSE = np.sqrt((1/len(section1))*(np.sum((section1 - section1_mean)**2)))
section2_RMSE = np.sqrt((1/len(section2))*(np.sum((section2 - section2_mean)**2)))
section3_RMSE = np.sqrt((1/len(section3))*(np.sum((section3 - section3_mean)**2)))
```

```python
# here are the RMSE values done for the given function
print(full_RMSE)
print(section1_RMSE)
print(section2_RMSE)
print(section3_RMSE)
```

```python
# now here we can compare these values to that recieved from sklearn
# print out the values
from sklearn.metrics import mean_squared_error

section1_mean_array = np.full(len(section1), section1_mean)
section2_mean_array = np.full(len(section2), section2_mean)
section3_mean_array = np.full(len(section3), section3_mean)

full_RMSE_sklearn = mean_squared_error(b, mean_array, squared=False)
section1_RMSE_sklearn = mean_squared_error(section1, section1_mean_array, squared=False)
section2_RMSE_sklearn = mean_squared_error(section2, section2_mean_array, squared=False)
section3_RMSE_sklearn = mean_squared_error(section3, section3_mean_array, squared=False)
print(full_RMSE_sklearn)
print(section1_RMSE_sklearn)
print(section2_RMSE_sklearn)
print(section3_RMSE_sklearn)
```

```python
# looks like sklearn implementation does a pretty good job

print(full_RMSE/full_RMSE_sklearn)
print(section1_RMSE/section1_RMSE_sklearn)
print(section2_RMSE/section2_RMSE_sklearn)
print(section3_RMSE/section3_RMSE_sklearn)
```
